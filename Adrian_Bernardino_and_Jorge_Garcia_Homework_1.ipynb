**Activation Function: Identity
g(x) = x** 



```python
x = np.arange(-5, 5, 0.01)
y = x
plt.plot(x,y)
plt.title('Identity')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
plt.show()
```

**Activation Function: Binary Step g(x) < 0 ; g(x) >= 0**

```python
def binaryStep(x):
  return np.heaviside(x, 1)

x = np.arange(-5, 5, 0.01)
y = x
plt.plot(x, binaryStep(y))
plt.title('Binary Step Activation Function')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
plt.show()
```

```python
def binaryStep(x):
  if x < 0:
    return 0
  else:
    return 1

binaryStep(8), binaryStep(-1)
```

**Activation Function: Sigmoid g(x) = 1/(1 + exp(-x))**

```python
def sigmoid(x):
  return 1/(1 + np.exp(-x))

x = np.arange(-5, 5, 0.01)
plt.plot(x, sigmoid(x))
plt.title('Sigmoid Activation Function')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
plt.show()
```

```python
def sigmoid(x):
  z = (1/(1 + np.exp(-x)))
  return z

sigmoid(4), sigmoid(-10)
```

**Activation Function: Hyperbolic Tangent g(x) = (exp(x) - exp(-x) / exp(x) + exp(-x))**

```python
def tanh(x):
  return np.tanh(x)

x = np.arange(-5, 5, 0.01)
plt.plot(x, tanh(x))
plt.title('Hyperbolic Tangent (tanh)')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
plt.show()
```

```python
def tanh(x):
  z = (2/(1 + np.exp(-2*x))) -1
  return z

tanh(-1), tanh(1)
```

**Activation Function: Recitfied linear unit g(x) = max(0,x)**

```python
def ReLU(x):
  x1 = []
  for i in x:
    if i < 0:
      x1.append(0)
    else:
      x1.append(i)
  return x1

x = np.arange(-5, 5, 0.01)
plt.plot(x, ReLU(x))
plt.title('Rectified linear unit (ReLU)')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
plt.show()
```

```python
def ReLU(x):
  if x < 0:
    return 0
  else: 
    return x

ReLU(-10), ReLU(10)
```

**Activation Function: Gaussian Error linear unit (GeLU) g(x) = 1/2x(1 + erf(x/sqrt(2))**

```python
from scipy.stats import norm
def GeLU(x):
  return x*norm.cdf(x)

def GeLU_approx(x):
  return .5 * x * 1(1 + tanh(np.sqrt(2/np.pi)*(x + 0.044715 * x**3)))

x = np.arange(-5, 5, 0.01)
plt.plot(x, GeLU(x))
plt.title('Gaussian Error linear unit (GeLU)')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
plt.show()
```

**Activation Function: Softplus g(x) = ln(1 + e(exp(-x)))**

```python
def softplus(x):
  return np.log(1 + np.exp(-np.abs(x))) + np.maximum(x, 0)

x = np.arange(-5, 5, 0.01)
plt.plot(x, softplus(x))
plt.title('Softplus')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
plt.show()
```

**Activation Function: Exponential Linear Unit g(x) = alpha (e^x - 1) if x <= 0 || g(x) = x if x > 0**

```python
def ELU(x, u):
  if x >= 0:
    return x
  else:
    return u*(np.exp(x) - 1)

eLU = np.vectorize(ELU)
x = np.arange(-5, 5, 0.01)
plt.plot(x, eLU(x, 1))
plt.title('Exponential Linear unit (ELU)')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
plt.show()
```

**Activation Function: Scaled Exponential Linear unit g(x) = lambda*x if x > 0 || g(x) = lambda * alpha(e^x - 1) if x <= 0**


```python
def SELU(x, lambdaa = 1.0507, alpha = 1.6732):
  if x >= 0:
    return lambdaa * x
  else:
    return lambdaa * alpha * (np.exp(x) - 1)

input = [x for x in range(-5, 10)]
output = [SELU(x) for x in input]
x = np.arange(-5, 5, 0.01)
plt.plot(input, output)
plt.title('Scaled Exponential Linear Unit (SeLU)')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
plt.show()  
```

**Activation Function: Leaky Rectified Linear Unit (Leaky ReLU) g(x) = 0.01x if x < 0 || g(x) = x if x >= 0**

```python
def lrelu(x):
  return x*0.01 if x < 0 else x

leaky = np.vectorize(lrelu)
x = np.arange(-10, 10, 0.01)
plt.plot(x, leaky(x))
plt.title('Leaky Rectified Linear Unit')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
plt.show()
```

**Activation Function: Parametric Rectified Linear Unit (PReLU) g(x) = alpha(x) if x < 0 || g(x) = x if x >= 0 with parameter alpha** 

```python
def prelu(x, alpha):
  if x >= 0: return x
  else: return alpha * x

preLu = np.vectorize(prelu)
x = np.arange(-10, 10, 0.01)
plt.plot(x, preLu(x, 0.01))
plt.title('Parametric Rectified Linear Unit')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
plt.show()
```

**Activation Function: Sigmoid Linear Unit (SiLU) g(x) = x/1 + e^-x** 

```python
def SiLU(x):
  return x / (1 + np.e**(-x))

x = np.arange(-5, 5, 0.01)
plt.plot(x, SiLU(x))
plt.title('SiLU')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
plt.show()
```

**Activation Function: Gaussian g(x) = e^-x^2**

```python
def Gaussian(x):
  return np.e**(-(x)**2)

x = np.arange(-5, 5, 0.01)
plt.plot(x, Gaussian(x))
plt.title('Gaussian')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
plt.show()
```
